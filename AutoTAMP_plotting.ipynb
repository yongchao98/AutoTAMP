# Figure for GPT-4 failure case

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_square(ax, x_start, x_end, y_start, y_end, color):
    ax.fill_between(np.linspace(x_start, x_end), y_start, y_end, color=color, alpha=1)

def draw_trajectory(ax, positions, line_style, color):
    for p1, p2 in zip(positions[:-1], positions[1:]):
        ax.plot([p1[0], p2[0]], [p1[2], p2[2]], line_style, color=color, linewidth=2)

# Draw environment
fig, ax = plt.subplots()

draw_square(ax, -1.5, -1.4, -1.5, 1.5, 'black')  # Black Wall 1
draw_square(ax, 1.4, 1.5, -1.5, 1.5, 'black')  # Black Wall 2
draw_square(ax, -1.5, 1.5, -1.5, -1.4, 'black')  # Black Wall 3
draw_square(ax, -1.5, 1.5, 1.4, 1.5, 'black')  # Black Wall 4

draw_square(ax, -1, -0.7, -0.25, 0.5, 'yellow')  # Yellow Box
draw_square(ax, 0, 0.9, -1, -0.5, 'red')  # Red Box
draw_square(ax, 0.2, 0.7, 0.8, 1.2, 'green')  # Green Box
draw_square(ax, -0.4, 0.4, -0.4, 0.4, 'blue')  # Blue Box

# Draw trajectory
trajectory_positions_1 = [
    (-1.3, -1, -1.3, -1),
    (-0.95, -0.7, -0.2, 0.45),
    (0.45, -0.95, -0.95, -0.55),
    (0.25, 0.95, 0.85, 1.15),
    (1.3, 1, 1.3, 1)
]

trajectory_positions_2 = [
    (-1.3, -1, -1.3, -1),
    (-0.95, -0.7, -0.2, 0.45),
    (0.45, -0.75, -0.95, -0.55),
    (0.45, -0.95, -0.95, -0.55),
    (-0.3, -0.05, 0.55, 0.75),
    (0.25, 0.95, 0.85, 1.15),
    (1.3, 1, 1.3, 1)
]

trajectory_positions_3 = [
    (-1.3, -1, -1.3, -1),
    (-0.95, -0.7, -0.2, 0.45),
    (-0.6, -0.4, 0.4, 0.6),
    (0.6, 0.4, 0.4, 0.6),
    (0.45, -0.95, -0.95, -0.55),
    (0.25, 0.95, 0.85, 1.15),
    (1.3, 1, 1.3, 1)
]

draw_trajectory(ax, trajectory_positions_1, 'o--', 'purple')
draw_trajectory(ax, trajectory_positions_2, '^--', 'gray')
#draw_trajectory(ax, trajectory_positions_3, '-', 'brown')
ax.plot(1.3, 1.3, '*', color = 'black')
ax.plot(-1.3, -1.3, 'o', color = 'black')

ax.set_xlim(-1.6, 1.6)
ax.set_ylim(-1.6, 1.6)
ax.set_aspect('equal', 'box')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Robot Trajectory")
plt.show()
```
